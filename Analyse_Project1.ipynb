```python
# Deal Creation Trend (Monthly): Group the deals data by month using the created_time column and count the number of deals.
deals['created_time'] = pd.to_datetime(deals['created_time'], errors='coerce')
deals['created_month'] = deals['created_time'].dt.to_period('M')

monthly_created_deals = (
    deals.groupby('created_month')
    .size()
    .reset_index(name='Created Deals Count (Monthly)')
)

monthly_created_deals['created_month'] = monthly_created_deals['created_month'].astype(str)

print(f'monthly_created_deals:')
display(monthly_created_deals.head())
```

```python
# Aggregate spend data by campaign
spend_agg = (
    spend.groupby('campaign')
    .agg(
        total_spend=('spend', 'sum'),
        total_clicks=('clicks', 'sum'),
        total_impressions=('impressions', 'sum')
    )
    .reset_index()
)

# Aggregate deals data by campaign
campaign_perf = (
    deals.groupby('campaign')
    .agg(
        total_deals=('id', 'count'),
        won_deals=('stage', lambda x: (x == 'Payment Done').sum())
    )
    .reset_index()
)

# Calculate conversion rate
campaign_perf['conversion_rate'] = (
    campaign_perf['won_deals'] / campaign_perf['total_deals']
)

# Merge campaign performance with spend data
campaign_perf_with_spend = (
    pd.merge(campaign_perf, spend_agg, on='campaign', how='left')
    .fillna(0)
)

# Calculate Cost Per Won Deal (CPW) and Click-Through Rate (CTR)
campaign_perf_with_spend['CPW'] = (
    campaign_perf_with_spend['total_spend'] / campaign_perf_with_spend['won_deals']
)
campaign_perf_with_spend['CTR'] = (
    campaign_perf_with_spend['total_clicks'] / campaign_perf_with_spend['total_impressions'] * 100
)

# Replace infinite values with NaN
campaign_perf_with_spend[['CPW', 'CTR']] = campaign_perf_with_spend[['CPW', 'CTR']].replace([np.inf, -np.inf], np.nan)
```

```python
# Aggregate deals data by source
source_deals_analysis = (
    deals.groupby('source')
    .agg(
        total_deals=('id', 'count'),
        won_deals=('stage', lambda x: (x == 'Payment Done').sum())
    )
    .reset_index()
)

# Aggregate spend data by source
source_spend_analysis = (
    spend.groupby('source')
    .agg(
        total_spend=('spend', 'sum'),
        total_clicks=('clicks', 'sum'),
        total_impressions=('impressions', 'sum')
    )
    .reset_index()
)

# Merge deals and spend data on source
source_full_analysis = pd.merge(
    source_deals_analysis,
    source_spend_analysis,
    on='source',
    how='left'
).fillna(0)

# Calculate conversion rate
source_full_analysis['conversion_rate'] = (
    source_full_analysis['won_deals'] / source_full_analysis['total_deals']
)

# Calculate Cost Per Lead (CPL) and Cost Per Won Deal (CPW)
source_full_analysis['cpl'] = (
    source_full_analysis['total_spend'] / source_full_analysis['total_deals']
)
source_full_analysis['cpw'] = (
    source_full_analysis['total_spend'] / source_full_analysis['won_deals']
)

# Replace infinite values with NaN and handle zero CPL and CPW
source_full_analysis[['cpl', 'cpw']] = source_full_analysis[
    ['cpl', 'cpw']
].replace([np.inf, -np.inf], np.nan)

source_full_analysis['cpl'].fillna(0, inplace=True)

source_full_analysis.loc[source_full_analysis['cpw'] == 0, 'cpw'] = np.nan
```

```python
columns_needed = ["city", "stage", "level_of_deutsch_standardized", "product"]

deals_analysis_df = deals.copy()

# Filter only required columns
deals_analysis_df = deals_analysis_df[
    [col for col in columns_needed if col in deals_analysis_df.columns]
].copy()

# Drop rows where city is missing
deals_analysis_df = deals_analysis_df.dropna(subset=["city"])

# Mark successful deals
deals_analysis_df["is_success"] = deals_analysis_df["stage"].apply(
    lambda x: 1 if str(x).strip().lower() == "payment done" else 0
)

# === Geographical distribution of transactions ===
city_summary = (
    deals_analysis_df
    .groupby("city", as_index=False)
    .agg(
        total_deals=("stage", "count"),
        successful=("is_success", "sum")
    )
)

city_summary["success_rate"] = (
    city_summary["successful"] / city_summary["total_deals"] * 100
).round(2)

print("    :")
print(city_summary.sort_values("total_deals", ascending=False).head())
```

```python
city_summary_filtered = city_summary[
    ~city_summary["city"].isin(["Unknown", "-"])
].copy()

fig2 = px.bar(
    city_summary_filtered.sort_values("total_deals", ascending=False).head(20),  # Top 20 cities
    x="city",
    y="total_deals",
    color="total_deals",
    text="total_deals",
    color_continuous_scale=px.colors.sequential.Blues,
    title="    ( 'Unknown'  '-')"
)

fig2.update_layout(template="plotly_white", xaxis_tickangle=45)
fig2.show()
```
